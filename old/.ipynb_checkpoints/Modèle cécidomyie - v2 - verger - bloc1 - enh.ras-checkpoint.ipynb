### Loi de développement du système

**Soit $N_t$ la population estimée d'adultes dans le verger à la date t.** Pour simuler la dynamique de cette population, on utilise l'équation suivante:

$$N_{t+1} = \lambda + L_{t+1-d_p} \times \mu_{MS} \times p_{pup} \times p_{l} $$

avec $\mu_{MS} = \mu^1_{MS} \times \mu^2_{MS}$.

Cette équation signifie que la population d'adultes dans le verger à la date $t+1$ et à la date $t-1$ est le nombre d'adultes entrant dans le verger ($\lambda$), plus le nombre de larves qui émergent du sol entrées en pupaison à la date $t+1-d_p$.

* $\lambda_t$ est le nombre d'adultes entrant dans le verger à la date t. Pour l'instant c'est un nombre constant au cours du temps appelé $\lambda$.

* $\mu^1_{MS}$ est la probabilité de survie de la larve de l’éjection au sol jusqu’à ce qu’elle rentre en terre selon la modalité du sol.

* $\mu^2_{MS}$ est la probabilité pour l’adulte de sortir du sol selon la modalité du sol.

* $p_{l}$ est la probabilité pour l’adulte de rester dans le verger quand il sort du sol. 

*  $d_p$ est la durée de pupaison. Entre 4 et 6 jours.

**Soit $L_t$ le nombre estimé de larves qui s'éjectent à la date t.**

$$L_{t} = N_{t-d_l} \times R \times E \times \mu$$

* $E$ est le nombre maximal d'oeufs pondus par une femelle.
* $\mu$ est la probabilité de survie des oeufs jusqu'au troisième stade de larve.
* $R$ la disponibilité en ressource (inflorescences).
* $d_l$ est la durée de dévelopement des larves et des oeufs. Entre 7 et 12 jours.


**Soit $L^p_t$ est le nombre estimé de larves piégées à la date t.**

$$L^p_t = L_t \times \beta$$ 

* $\beta$ est l'efficacité du piège.
   - Les pièges utilisés en 2017 sont différents de ceux utilisés dans l'expérience de Paul. Il faudrait réestimer l'efficacité de ce nouveau type de pièges.
   - La valeur du beta dans le modèle de Paul est > 1. Ce qui parait un peu abérant (plus de larves piégées que de larves qui tombent).

**Soit $I^d_t$ la population d'inflos à l'instant $t$ apparues le jour $d$.**

$$I^d_{t} = \begin{cases}
I^+(d) & \text{si } t = d \\
I^d_{t-1} (1 - \gamma \times N_{t-1}) & \text{si } d < t < d+T\\
0 & \text{sinon }
\end{cases}$$

 * $T$ la durée de vie d'une inflorescence.
 
Soit $I_t$ le nombre estimé d'inflos dans le verger.

$$I_t = \sum_d I^d_t$$

**Pour le moment, $I_t$ est une entrée du modèle. On le récupère des données de Christian et Alain**


On estime la disponibilité en ressource (inflorescence) par un coefficient $R$ estimé de la manière suivante:

$$R =  \begin{cases} 1 -  \frac{N_t}{k \times I_t} & \text{ si } N_t < k \times I_t \\ 0 & \text{ sinon. }  \end{cases} $$

 * $k$ est une constante. C'est le nombre de piqûre qui garantit la mort de l'inflo. 

Ca sert à ne pas compter les larves si elles ne peuvent pas se poser sur une inflo (i.e. s'il n'y a pas de ressources dans le verger)


#### Liste des paramètres recherchés

$\lambda_t$ est le nombre d'adultes entrant dans le verger à la date t. Pour l'instant c'est un nombre constant au cours du temps appelé $\lambda$.

$\mu^1_{MS}$ est la probabilité de survie de la larve de l’éjection au sol jusqu’à qu’elle rentre en terre selon la modalité du sol

$\mu^2_{MS}$ est la probabilité pour l’adulte de sortir du sol selon la modalité du sol

$p_{l}$ est la probabilité pour l’adulte de rester dans le verger quand il sort du sol selon la modalité du sol

$k$ est le nombre de piqûres qui garantit la mort de l'inflo

$\gamma$ est le taux de dégradation des inflos 

#### Liste des paramètres connus

$\bar{L_t}$ est le nombre observé de larves piégées à la date t.

$E$ est le nombre maximal d'oeufs pondus par une femelle.

($I^+$ est le nombre d'apparition d'inflos.)

$d_l$ est la durée de dévelopement des larves et des oeufs. Entre 7 et 12 jours.

$d_p$ est la durée de pupaison. Entre 4 et 6 jours.

$\beta$ est l'efficacité du piège. 

$p_{pup}$ est la probabilité d'entrer en pupaison.

$\mu$ est la probabilité de survie des oeufs jusqu'au troisième stade de larve.

$T$ est la durée de vie d'une inflo, estimée d'après Jestin.

```python
nbArbreTot = 49.
nbArbrePiege = 10.
coefArbre = nbArbreTot/nbArbrePiege

nb_larves = np.zeros(len(liste1)-1) 
nb_inflo_piege = np.zeros(len(liste1)-1) 
nb_inflo = np.zeros(len(liste1)-1) 
nb_inflo_morte = np.zeros(len(liste1)-1) 
nb_larves_estima = np.zeros(len(liste1)-1) 
#larve = np.zeros(len(liste1)-1)

for i in range (1,len(liste1)) :
    nb_larves[i-1] = liste1[i][2]
    nb_inflo_piege[i-1] = liste1[i][3]
    nb_inflo[i-1] = liste1[i][4]
    nb_inflo_morte[i-1] = liste1[i][5]
    #larve[i-1] = liste1[i][6]
    nb_larves_estima[i-1] = liste1[i][6]

# Nombre de larves piégées à l'échelle du verger
nb_larves = nb_larves*coefArbre
# Nombre d'inflos piégées à l'échelle du verger
nb_inflo_piege = nb_inflo_piege*coefArbre
# Nombre d'inflos vivantes à l'échelle du verger
nb_inflo = nb_inflo*coefArbre
# Nombre d'inflos mortes à l'échelle du verger
nb_inflo_morte = nb_inflo_morte*coefArbre
# Nombre de larves piégées à l'échelle du verger 
nb_larves_estima = nb_larves_estima*coefArbre
```

```python
# listeJours contient toutes les dates de relevés (jour, mois)
listeJours = np.array(((18,7),(25,7),(1,8),(8,8),(16,8),(18,8),(23,8),(25,8),(29,8),(1,9),(6,9),(8,9),(13,9),(15,9),(20,9),(22,9),(26,9),(29,9),(3,10),(6,10)))
NbJours = len(listeJours)

# Cette fonction renvoie le numéro de la date day en fonction d'une date de référence refday (le numéro de la date refday est 0)
def daynb (day, refday) :
    return (date(2017,day[1], day[0]) - date(2017,refday[1], refday[0])).days

# xJours contient les dates de relevés en "nombre"
xJours = list (map(lambda d : daynb(d,(18,7)), listeJours))

# TotJours est le nombre de jours entre le premier relevé et le dernier
TotJours = xJours[-1]
# xTotJours contient toutes les dates comprises entre le premier relevé et le dernier en "nombre"
xTotJours = np.linspace(0,TotJours,TotJours+1)
```

### Optimisation du système

```python
# Cette fonction renvoie chaque variable du système au temps t suivant. 

# It est d'abord mis en entrée.
It = nb_inflo_piege

def step (t, Nt, Lt, Lt_p, lambda_, p_l, k, gamma) :
    
    mu_MS = 1
    
    # R : coefficient estimant la disponibilité en ressource à la date t 
    # On suppose qu'à la date 1 toutes les ressources sont disponibles 
    if (t==1) : 
        R = 1
    else :
        if (Nt[-1] < (k*It[-1])):
            R = 1 - Nt[-1] / (k*It[-1])
        else :
            R = 0
    #print t, R
    # Lt_plus_1 est le nombre de larves qui s'ejectent à la date t 
    # Jusqu'à la date d_l+1, il n'y a pas de larves qui s'éjectent car d_l est le temps de développement des oeufs jusqu'au 
    # troisième stade de larve. Donc, les premières larves qui s'éjectent sont celles issues des oeufs posés par une femelle
    # adulte à la date 0. 
    if (t <= d_l) :
        Lt_plus_1 = 0
    else : 
        Lt_plus_1 = Nt[-1-d_l] * R * E * mu / 2 # On divise par 2 à cause du sex ratio
    
    # Lt_p_plus_1 est le nombre de larves récoltées par les pièges à la date t 
    Lt_p_plus_1 = Lt_plus_1 * beta
    
    # Nt_plus_1 est le nombre de larves dans le verger à la date t
    # Au début, on ne compte que les larves entrant dans le verger, ensuite, celles entrant dans le verger et celles 
    # émergeant du sol lorsqu'on est assez avancé dans le temps et qu'elles ont eu le temps de faire leur pupaison
    Nt_plus_1 = lambda_
    if (t > d_p) :
        Nt_plus_1 +=  Lt[-1-d_p] * mu_MS * p_pup * p_l
    
    # It_plus_1 est le nombre d'inflos au dessus des pièges à la date t
    # It_d[i] est le nombre d'inflo apparue au temps i et toujours présente à la date t-1, puis t 
    #It_tmp = np.zeros((TotJours+1))
    #It_tmp[t] = II[t]
    #for i in range (max(t-T,0),t) :
    #    It_tmp[i] = It_d[i]*(1-gamma*Nt[-1])
    #It_d = It_tmp
    #It_plus_1 = np.sum(It_d)
    
    # On ajoute les valeurs de la date t au vecteur 
    Nt.append(Nt_plus_1)
    #It.append(It_plus_1)
    Lt.append(Lt_plus_1)
    Lt_p.append(Lt_p_plus_1)
    
    return Nt, Lt, Lt_p
```

```python
# Cette fonction calcule toutes les variables du système au fil du temps et renvoie Lt_p, le nombre de larves piégées à chaque
# pas de temps. C'est sur ces données qu'on fait la minimisation.

def integrate (lambda_, p_l, k, gamma) :
    
    # Initialisation 
    Nt = [lambda_]
    Lt = [0]
    Lt_p = [0]
    #It = [II[0]]
    #It_d = (1/II[0])*np.ones((TotJours+1))
    
    for i in range (1,TotJours+1) :
        Nt, Lt, Lt_p = step (i, Nt, Lt, Lt_p, lambda_, p_l, k, gamma)
        
    # A la fin de la boucle, 
    # Nt : population de cécidomyie ...
    # It : nombre d'inflos au dessus des pièges ...
    # Lt : nombre de larves qui s'éjectent ...
    # Lt_p : nombre de larves piégées ... 
    # ... au cours du temps 
        
    return np.array(Lt), np.array(Lt_p), np.array(Nt) 
```

Deux fonction objectifs sont possibles, une minimisant sur le nombre de larves capturées : $\sum ||L_t - \bar{L_t}||$, l'autre sur le nombre d'inflos au dessus des pièges $\sum ||I_t - \bar{I_t}||$. On calcule la distance uniquement sur les valeurs vraiment observées et non celles qui sont interpolées. Les voici, ci-dessous. 

```python
distvalues = []

# Cette fonction minimise sur le nombre de larves piégées
def objectif1_1 (params) :
    global distvalues
    val = distance1(integrate(*params)[1][xJours], nb_larves_estima)
    distvalues.append(val)
    return val

def objectif1_2 (params) :
    global distvalues
    val = distance2(integrate(*params)[1][xJours], nb_larves_estima)
    distvalues.append(val)
    return val

def objectif1_Inf (params) :
    global distvalues
    val = distanceInf1(integrate(*params)[1][xJours], nb_larves_estima)
    distvalues.append(val)
    return val
```

```python
# Fonctions distance pour les fonctions objectifs 

# Distance 1
def distance1 (Simu,Obs) :
    return sum(abs(Simu-Obs))

# Distance euclidienne
def distance2 (Simu,Obs) :
    return np.sqrt(sum((Simu-Obs)**2))

# Distance inf 
def distanceInf1 (Simu,Obs) :
    return max(abs(Simu-Obs))

def distanceInf2 (Simu1,Obs1,Simu2,Obs2) :
    return max(max(abs(Simu1-Obs1)),max(abs(Simu2-Obs2)))
```

```python
# Bornes pour les paramètres à estimer
bounds = ([0, 0, 0, 0], [np.inf, 1, np.inf, 1])
```

```python
# Fonctions d'optimisation

def optimize1 (p0) :
    res = least_squares(objectif1_1, p0, bounds=bounds, method='trf')
    return res.x

def optimize2 (p0) :
    res = least_squares(objectif1_2, p0, bounds=bounds, method='trf')
    return res.x

def optimizeInf (p0) :
    res = least_squares(objectif1_Inf, p0, bounds=bounds, method='trf')
    return res.x
```

```python
# Fonction de visuaisation des résultats de toutes les optimisations 

def visual_optimization_bis (optimize, p0) :
    
    global distvalues
    
    min_ = 99999999999999999999999
    for i,j,kk,l in zip(p0[0],p0[1],p0[2],p0[3]):
        distvalues = []
        res = optimize((i,j,kk,l))
        if (min(distvalues)<min_):
            min_init = (i,j,kk,l)

    distvalues = []
    res = optimize((min_init[0], min_init[1], min_init[2], min_init[3]))
    L_simu, L_t_simu, N_simu = integrate (res[0], res[1], res[2], res[3])
    
    plt.plot(xJours,L_t_simu[xJours],'o-',label=u"Simulé")
    plt.plot(xJours,nb_larves_estima,'o-',label=u"Observé")
    plt.legend()
    plt.title(u"Nombre de cécidomyies au cours du temps dans le verger")
    plt.show()
    
    print (sum(L_t_simu[xJours]),sum(nb_larves_estima))
    
    plt.plot(np.log(distvalues))
    plt.legend()
    plt.title("Fonction objectif")
    plt.show()
    
    return min(distvalues), res, min_init, L_t_simu[xJours]
```
