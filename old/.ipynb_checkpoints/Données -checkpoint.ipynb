## Fichier piege

```r
# Lecture du fichier piege 
data = read.table(file = "piege.csv",header=TRUE,sep=";")
```

### Comment récupère-je les données? 

Une ligne du fichier piège contient le numéro de bloc (1 ou 2), le traitement du sol appliqué (bache, enherbement haut ou ras), le numéro d'arbre de la sous parcelle (Bloc-Traitement), la date de relevé, le nombre de larves récoltées par le piège, le nombre d'inflorescences vivantes au dessus du piège, le nombre d'inflorescences vivantes dans l'arbre, le nombre d'inflorescences mortes dans l'arbre, le nombre de larves estimées à l'échelle de l'arbre (je ne récupère pas cette information, mais refais le calcul car il n'est pas correct), le nombre de jours écoulés depuis le précédent relevé (je n'utilise pas cette information pour le moment), la proportion de larves piégées par rapport aux inflorescences au dessus du piège et des observations (je ne récupère pas cette information). 

Pour le moment, le modèle est non spatialisé, donc je vais récupérer toutes les données de chaque arbre de façon confondue en les ajoutant. L'idée est de mettre chaque donnée, d'abord à l'échelle de l'arbre, d'additionner toutes les données, puis, de les mettre à l'échelle du verger pour toute date de relevé.

Voici, comment je procède pour le bloc1 en entier (les 3 parcelles confondues) : 

#### Bloc 1, les 3 parcelles confondues, sans spatialisation

Pour mettre à l'échelle du verger, on a besoin du nombre d'arbres piégés dans le verger et du nombre total d'arbres dans le verger.

```r
nbArbrePiege = 30
nbArbreTot = 153
coefArbre = nbArbreTot/nbArbrePiege
```

Dans le fichier piege, je récupère toutes les données pour le bloc 1, je met les valeurs NA à 0, je trie par date de faon chronologique et je stocke le tout dans une variable B.

```r
B = subset(data[data$bloc==1,], 
           select=c(bloc, trait, arbre, piege, date, nb.larves, nb.inflo.piege, nb.inflo, nb.inflo.morte, nb.larves.estima, nb.jours, larve))
B = replace(B,is.na(B),0)
B = B[order(as.Date(B$date, '%d/%m/%Y')),]
```

Je récupère les dates de relevé.

```r
date = unique(B$date)
```

** 1. ** LES LARVES PIEGEES

Jusqu'au 16/08/2017 inclus, il n'y a qu'un piège A sous les arbres piégés. J'estime le nombre de larves par arbre piégé avec la formule suivante : $$\text{Nombre de larves récoltées par le piège A} \times \frac{\text{Nombre d'inflos dans l'arbre}}{\text{Nombre d'inflos au dessus du piège A}}$$ et je fais la somme de toutes ces larves pour chaque date de relevé. 

A partir du 18/08/2017, il y a deux pièges (A et B) sous les arbres piégés. J'estime le nombre de larves par arbre piégé avec la formule suivante : $$\text{Nombre de larves récoltées par les pièges A et B} \times \frac{\text{Nombre d'inflos dans l'arbre}}{\text{Nombre d'inflos au dessus des pièges A et B}}$$ et je fais la somme de toutes ces larves pour chaque date de relevé.

Enfin, je multiplie par le coefficient $\text{coefArbre}$ pour passer à l'échelle du verger.

```r
larves = array(0, length(date))

for (i in 1:length(date)) {
    tmp = B[B$date==date[i],]
    j = 1
    while (j<dim(tmp)[1]) {
        if (tmp[j,]$trait==tmp[j+1,]$trait & tmp[j,]$arbre==tmp[j+1,]$arbre) {
            if (tmp$nb.inflo.piege[j]+tmp$nb.inflo.piege[j+1]!=0)
                larves[i] = larves[i] + (tmp$nb.larves[j]+tmp$nb.larves[j+1])*tmp$nb.inflo[j]/(tmp$nb.inflo.piege[j]+tmp$nb.inflo.piege[j+1])
            j = j+2
        }
        else {
            if (tmp$nb.inflo.piege[j]!=0)
                larves[i] = larves[i] + tmp$nb.larves[j]*tmp$nb.inflo[j]/tmp$nb.inflo.piege[j]
            j = j+1
        }
    }
}

larves = larves * coefArbre
```

** 2. ** LES INFLORESCENCES VIVANTES ET MORTES

Je récupère le nombre d'inflorescences vivantes et mortes dans chaque arbre piégé en additionnant le nombre d'inflorescences vivantes de chaque arbre pour le piège A. 

Enfin, je multiplie par le coefficient $\text{coefArbre}$ pour passer à l'échelle du verger.

```r
inflos_vivantes = array(0, length(date))
inflos_mortes = array(0, length(date))

BA = B[B$piege=='A',][order(as.Date(B[B$piege=='A',]$date, '%d/%m/%Y')),]
BA = replace(BA,is.na(BA),0)

for (i in 1:length(date)) {
    tmp = BA[BA$date==date[i],]
    inflos_vivantes[i] = sum(tmp$nb.inflo)
    inflos_mortes[i] = sum(tmp$nb.inflo.morte)
}

inflos_vivantes = inflos_vivantes * coefArbre
inflos_mortes = inflos_mortes * coefArbre
```

** 3. ** LE NOMBRE DE JOURS ENTRE LES RELEVES

Je récupère le nombre de jours entre chaque relevé en prenan ceux pour le premier piège.

```r
nb_jours = B[B$trait=='bache' & B$arbre==2 & B$piege=='A', ]$nb.jours
nb_jours[1] = 0
```

Je met les données récoltées dans un seul dataframe, puis, dans un fichier excel.

```r
Bloc1 = data.frame(date, larves, inflos_vivantes, inflos_mortes, nb_jours)
write.csv(Bloc1,"Bloc1.csv")
```

## Fichier Dynamique cecido modif

```r
data1 = read.table(file = "dynamique.cecido.modif.csv",header=TRUE,sep=";")
```

```r
# On enlève les stades phénos non observés
data1 = data1[!is.na(data1$stade),]

# On sépare selon le bloc, le traitement du sol, l'arbre, le piège, et le numéro d'étiquette afin d'avoir toutes les 
# étiquettes suivies séparées 
data1_split = split(data1, list(data1$bloc,data1$trait,data1$arbre,data1$piege,data1$num.etiq))

# Nombre d'étiquettes suivies avec des stades phénos observés
NbGroup = length(data1_split)

# Pour la première étiquette :
    # On récupère la date de naissance de l'inflo (premier stade phéno observé)
    # On récupère la date de mort de l'inflo (premier 5 dans le niveau de déssechement, signifie qu'elle est sèche à 100%)
tmp = data.frame(data1_split[2])
colnames(tmp) = colnames(data1)
tmp$date = as.Date(tmp$date, '%d/%m/%Y')
birth <- c(tmp$date[1])
j = min(which(tmp$etat.panicule == 5))
dead = c(strptime(tmp$date[j], format = "%Y-%m-%d"))
if (1 %in% tmp$cecido) {
  cecido = c(1)
} else {
    cecido = c(0)
}

# Pour les autres étiquettes :
    # On récupère la date de naissance de l'inflo (premier stade phéno observé)
    # On récupère la date de mort de l'inflo (premier 5 dans le niveau de déssechement, signifie qu'elle est sèche à 100%)
for (i in 3:NbGroup) {
  tmp = data.frame(data1_split[i])
  colnames(tmp) = colnames(data1)
  tmp$date = as.Date(tmp$date, '%d/%m/%Y')
  if (dim(tmp)[1]!=0 & 5 %in% tmp$etat.panicule) {
    birth = c(birth, tmp$date[1])
    j = min(which(tmp$etat.panicule == 5))
    dead = c(dead,tmp$date[j])
    if (1 %in% tmp$cecido) {
      cecido = c(cecido,1)
    }
    else
      cecido = c(cecido,0)
  }
}

datebirth = strptime(birth, format = "%Y-%m-%d")
datedead = strptime(dead, format = "%Y-%m-%d")
diff = as.double(difftime(dead,birth, units = "days"))
```
