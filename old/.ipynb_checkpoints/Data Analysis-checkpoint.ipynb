```python
from pandas import *

df = read_excel('piege0.xlsx','piege')
```

```python
from matplotlib.pyplot import *

df2 = df.groupby('date').sum()
#print df2
plot(df2.index,df2['nb.inflo'])
plot(df2.index,df2['nb.inflo.morte'])
plot(df2.index,df2['nb.inflo']+df2['nb.inflo.morte'])
xticks(rotation=60)
show()
```

```python
from matplotlib.pyplot import *

df2 = df.groupby('date').sum()
#print df2
plot(df2.index,df2['nb.inflo.morte']/(df2['nb.inflo']+df2['nb.inflo.morte']))
show()
```

```python
data = df2['nb.inflo.morte']/(df2['nb.inflo']+df2['nb.inflo.morte'])
values = data.values
from numpy import *
plot(log(values))
```
