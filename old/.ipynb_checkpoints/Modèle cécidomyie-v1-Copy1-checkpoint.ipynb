### Importation des librairies

```python
from scipy.optimize import least_squares
import numpy as np
import csv
import random
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.stats import norm
import scipy
from datetime import date, timedelta
```

### Loi de dévelopement du système

Soit $N_t$ la population estimée d'adultes dans le verger à la date t.

$\frac{dN_t}{dt} = \lambda + L_t \times \mu_{MS} \times p_{pup} \times p_{l} - \alpha * N_{t-1}$.

avec $\alpha$ le taux de mortalité égal à 1 et $\mu_{MS} = \mu^1_{MS} \times \mu^2_{MS}$.

Cette équation signifie que la différence entre la population d'adultes dans le verger à la date $t$ et à la date $t-1$ est le nombre d'adultes entrant dans le verger ($\lambda$), plus le nombre de larves qui émergent du sol entrées en pupaison à la date $t-d_p$ moins la population à la date $t-1$.

Soit $L_t$ le nombre estimé de larves qui s'éjectent à la date t.

$L_t = N_{t-d_l} \times R \times E \times \mu$. 

Soit $L^p_t$ est le nombre estimé de larves piégées à la date t.

$L^p_t = L_t \times \beta$. 

Soit $I_t$ le nombre estimé d'inflos dans le verger.

$\frac{dI_t}{dt} = - I_{t-1} \times \gamma \times N_t + I^+$

On estime la disponibilité en ressource (inflorescence) par un coefficient $R$ estimé de la manière suivante:

$R = 1 -  \frac{N_t}{k \times I_t}$ si $N_t < k \times I_t$, 0 sinon. 

$k$ est une constante. C'est le nombre de piqûre qui garantit la mort de l'inflo. 

Ca sert à ne pas compter les larves si elles ne peuvent pas se poser sur une inflo (i.e. s'il n'y a pas de ressources dans le verger)


On utilise les différences finies sur les équations suivantes : $\frac{dI_t}{dt} = - I_{t-1} \times \gamma \times N_t + I^+$ et $\frac{dN_t}{dt} = \lambda + L_{t-d_p} \times \mu_{MS} \times p_{pup} \times p_{l} - \alpha * N_{t-1}$.

Ce qui donne : $\frac{I_t - I_{t-1}}{\Delta t} \approx- I_{t-1} \times \gamma \times N_t + I^+$ et $\frac{N_t - N_{t-1}}{\Delta t} = \lambda + L_{t-d_p} \times \mu_{MS} \times p_{pup} \times p_{l} - \alpha * N_{t-1}$.

Enfin : $I_t \approx I_{t-1}(1 - \gamma \times N_t) + I^+$ car $\Delta t = 1$ et $N_t \approx \lambda + L_{t-d_p} \times \mu_{MS} \times p_{pup} \times p_{l} + (1 - \alpha)N_{t-1}$. (On suppose qu'il y a une semaine, une unité de temps, entre chaque relevé)

#### Liste des paramètres recherchés

$\lambda_t$ est le nombre d'adultes entrant dans le verger à la date t. Pour l'instant c'est un nombre constant au cours du temps.

$\mu^1_{MS}$ est la probabilité de survie de la larve de l’éjection au sol jusqu’à qu’elle rentre en terre selon la modalité du sol

$\mu^2_{MS}$ est la probabilité pour l’adulte de sortir du sol selon la modalité du sol

$p_{l}$ est la probabilité pour l’adulte de rester dans le verger quand il sort du sol selon la modalité du sol

$\gamma$ est le taux de dégradation des inflos 

#### Liste des paramètres connus

$\bar{L_t}$ est le nombre observé de larves piégées à la date t.

$E$ est le nombre maximal d'oeufs pondus par une femelle.

$I^+$ est le nombre d'apparition d'inflos. En première approximation, suit une loi gaussienne sur la période de floraison.

$d_l$ est la durée de dévelopement des larves et des oeufs. Entre 7 et 12 jours.

$d_p$ est la durée de pupaison. Entre 4 et 6 jours.

$\beta$ est l'efficacité du piège. 

$p_{pup}$ est la probabilité d'entrer en pupaison.

$\mu$ est la probabilité de survie des oeufs jusqu'au troisième stade de larve.

$k$ est le nombre de piqûres qui garantit la mort de l'inflo. On le met 20 d'après le fichier dynamique.cecido.modif0.xlsx.

$\alpha$ le taux de mortalité égal à 1.

```python
E = 150
mu = 0.04
d_l = 10
d_p = 5
beta = np.mean(np.array((1,1,1.07,0.76,0.84,1.83))) # = 1.083
p_pup = (71.2+77)/2/100 # = 0.741
k = 20
alpha = 1
```

```python
# On récupère dans les fichiers de données le nombre de larves récoltées par tous les pièges à chaque date de relevé
# On stocke ces valeurs dans L_obs

f1 = open('Bloc1_nb.larves.csv', 'r')
test1 = csv.reader(f1)
liste1 = []
for row in test1:
    liste1.append(row)
f1.close()

L_obs = np.zeros(len(liste1)-1)
for i in range (1,len(liste1)) :
    L_obs[i-1] = liste1[i][2]
```

```python
# On récupère dans les fichiers de données le nombre d'inflos au dessus de tous les pièges à chaque date de relevé
# On stocke ces valeurs dans I_obs

f2 = open('Bloc1_nb.inflo.piege.csv', 'r')
test2 = csv.reader(f2)
liste2 = []
for row in test2:
    liste2.append(row)
f2.close()
    
I_obs = np.zeros(len(liste2)-1)
for i in range (1,len(liste2)) :
    I_obs[i-1] = liste2[i][2]
```

```python
# listeJours contient toutes les dates de relevés (jour, mois)
listeJours = np.array(((18,7),(25,7),(1,8),(8,8),(16,8),(18,8),(23,8),(25,8),(29,8),(1,9),(6,9),(8,9),(13,9),(15,9),(20,9),(22,9),(26,9),(29,9),(3,10),(6,10)))
NbJours = len(listeJours)

# Cette fonction renvoie le numéro de la date day en fonction d'une date de référence refday (le numéro de la date refday est 0)
def daynb (day, refday) :
    return (date(2017,day[1], day[0]) - date(2017,refday[1], refday[0])).days

# xJours contient les dates de relevés en "nombre"
xJours = list (map(lambda d : daynb(d,(18,7)), listeJours))

# TotJours est le nombre de jours entre le premier relevé et le dernier
TotJours = xJours[-1]
# xTotJours contient toutes les dates comprises entre le premier relevé et le dernier en "nombre"
xTotJours = np.linspace(0,TotJours,TotJours+1)
```

```python
# On affiche le nombre d'inflos au dessus des pièges à chaque date t, données observées 

plt.plot(xJours,I_obs,'o-',label=u"Observé")
plt.legend()
plt.title(u"OBSERVATION : Nombre d'inflos au dessus des pièges à chaque date t")
plt.show()
```

```python
# Pour le moment, on approxime le nombre d'inflos qui apparaissent à chaque date t avec la loi gaussienne suivante
# On stocke ces valeurs dans II 

# Allure de notre gaussienne 
xx = np.linspace(0,TotJours,TotJours+1)

II = np.floor(scipy.stats.norm.pdf(np.linspace(0,TotJours,TotJours+1),60,30)*sum(I_obs)*1.2*TotJours/len(I_obs))
plt.plot(xx,II)
plt.plot(xJours,I_obs,'o-',label=u"Observé")
plt.show()
```

```python
# On affiche le nombre d'inflos qui apparaissent au dessus des pièges à chaque date t (i.e. la différence entre le nombre 
# d'inflos à la date t et celles à la date t-1).
# Quand il y a plus d'inflos à la date t-1 qu'à la date t, il n'y pas d'apparition d'inflos, donc on met la valeur à 0 pour ne
# pas avoir une valeur négative.

xx = xJours[1:]
L_diff_obs = np.zeros(NbJours-1)
for i in range (NbJours-1) :
    L_diff_obs[i] = L_obs[i+1]-L_obs[i]
    if (L_diff_obs[i] < 0) :
        L_diff_obs[i] = 0
plt.plot(xx,L_diff_obs,'o-',label=u"Observé")
plt.legend()
plt.title(u"OBSERVATION : Nombre d'inflos qui apparaissent au dessus des pièges à chaque date t")
plt.show()
```

#### Interpolation des valeurs observées 

```python
# On interpole avec une fonction linéaire les valeurs observées pour le nombre d'inflos. 
# Ainsi, nous avons une aproximation du nombre d'inflos pour chaque jour, entre le premier et le dernier relevé que l'on
# stocke dans I_inter. 
f1 = interp1d(xJours, I_obs, kind='linear')
I_inter = f1(xTotJours)

# Affichage de l'interpolation
plt.plot(xJours, I_obs, 'o', label=u'Observé')
plt.plot(xTotJours, I_inter, '-',label=u'Interpolé')
plt.legend()
plt.title(u"Interpolation de la fonction pour le nombre d'inflos à chaque date t")
plt.show()
```

```python
# On interpole avec une fonction linéaire les valeurs observées pour le nombre de larves piégées. 
# Ainsi, nous avons une aproximation du nombre de larves piégées pour chaque jour, entre le premier et le dernier relevé que 
# l'on stocke dans I_inter. 
f2 = interp1d(xJours, L_obs, kind='linear')
L_inter = f2(xTotJours)

# Affichage de l'interpolation
plt.plot(xJours, L_obs, 'o', label=u'Observé')
plt.plot(xTotJours, L_inter, '-',label=u'Interpolé')
plt.legend()
plt.title(u"Interpolation de la fonction pour le nombre de larves piégées à chaque date t")
plt.show()
```

### Optimisation du système

```python
# Cette fonction renvoie chaque variable du système au temps t suivant. 

def step (t, Nt, It, Lt, Lt_p, lambda_, mu_MS, pl, k) :
    
    # R : coefficient estimant la disponibilité en ressource à la date t 
    # On suppose qu'à la date 1 toutes les ressources sont disponibles 
    if (t==1) : 
        R = 1
    else :
        R = max(0,1 - Nt[-1] / (k*It[-1]))
    
    # Lt_plus_1 est le nombre de larves qui s'ejectent à la date t 
    # Jusqu'à la date d_l+1, il n'y a pas de larves qui s'éjectent car d_l est le temps de développement des oeufs jusqu'au 
    # troisième stade de larve. Donc, les premières larves qui s'éjectent sont celles issues des oeufs posés par une femelle
    # adulte à la date 0. 
    if (t <= d_l) :
        Lt_plus_1 = 0
    else : 
        Lt_plus_1 = Nt[-1-d_l] * R * E * mu
    
    # Lt_p_plus_1 est le nombre de larves récoltées par les pièges à la date t 
    Lt_p_plus_1 = Lt_plus_1 * beta
    
    # Nt_plus_1 est le nombre de larves dans le verger à la date t
    # Au début, on ne compte que les larves entrant dans le verger, ensuite, celles entrant dans le verger et celles 
    # émergeant du sol lorsqu'on est assez avancé dans le temps et qu'elles ont eu le temps de faire leur pupaison
    Nt_plus_1 = lambda_ + (1-alpha) * Nt[-1]
    if (t > d_p) :
        Nt_plus_1 +=  Lt[-1-d_p] * mu_MS * p_pup * pl 
    
    # It_plus_1 est le nombre d'inflos au dessus des pièges à la date t    
    It_plus_1 = It[-1]*(1-min(1,Nt[-1]/(k*It[-1]))) + II[t]
    
    # On ajoute les valeurs de la date t au vecteur 
    Nt.append(Nt_plus_1)
    It.append(It_plus_1)
    Lt.append(Lt_plus_1)
    Lt_p.append(Lt_p_plus_1)
    
    return Nt, It, Lt, Lt_p
```

```python
# Cette fonction calcule toutes les variables du système au fil du temps et renvoie Lt_p, le nombre de larves piégées à chaque
# pas de temps. C'est sur ces données qu'on fait la minimisation.

def integrate (lambda_, mu_MS, pl, k) :
    
    # Initialisation 
    Nt = [lambda_]
    Lt = [0]
    Lt_p = [0]
    It = [II[0]]
    
    for i in range (1,TotJours+1) :
        Nt, It, Lt, Lt_p = step (i, Nt, It, Lt, Lt_p, lambda_, mu_MS, pl, k)
        
    # A la fin de la boucle, 
    # Nt : population de cécidomyie ...
    # It : nombre d'inflos au dessus des pièges ...
    # Lt : nombre de larves qui s'éjectent ...
    # Lt_p : nombre de larves piégées ... 
    # ... au cours du temps 
        
    return np.array(Lt_p), np.array(It) , np.array(Nt) 
```

Deux fonction objectifs sont possibles, une minimisant sur le nombre de larves capturées : $\sum ||L_t - \bar{L_t}||$, l'autre sur le nombre d'inflos au dessus des pièges $\sum ||I_t - \bar{I_t}||$. On calcule la distance uniquement sur les valeurs vraiment observées et non celles qui sont interpolées. Les voici, ci-dessous. 

```python
distvalues = []
def objectif1 (params) :
    global distvalues
    val = distance(integrate(*params)[0][xJours], L_obs)
    distvalues.append(val)
    return val

def objectif2 (params) :
    return distance(integrate(*params)[1][xJours], I_obs)
```

```python
# Fonction distance pour la fonction objectif 

def distance (Simu,Obs) :
    return sum((Simu-Obs)**2)
```

```python
# Fonctions d'optimisation 
bounds = ([0, 0, 0, 0], [np.inf, 1, 1, 1000])

def optimize1 (p0) :
    res = least_squares(objectif1, p0, bounds=bounds,method='dogbox')
    return res.x

def optimize2 (p0) :
    res = least_squares(objectif1, p0, bounds=bounds,method='trf')
    return res.x

def optimize4 (p0) :
    res = least_squares(objectif2, p0, bounds=bounds,method='dogbox')
    return res.x

def optimize5 (p0) :
    res = least_squares(objectif2, p0, bounds=bounds,method='trf')
    return res.x
```

```python
# Fonction de visualisation des résultats de l'optimisation 

def visual_optimization (optimize, p0) :
    
    assert len(xJours) == len(L_obs)
    
    global distvalues
    distvalues = []

    res = optimize(p0)
    
    L_simu, I_simu, N_simu = integrate (res[0], res[1], res[2], res[3])
    
    plt.plot(range(xJours[0],xJours[-1]+1),L_simu,'o-',label=u"Simulé")
    plt.plot(xJours,L_obs,'o-',label=u"Observé")
    plt.legend()
    plt.title(u"Nombre de larves au cours du temps")
    plt.show()
    
    plt.plot(range(xJours[0],xJours[-1]+1),N_simu,'o-',label=u"Simulé")
    plt.legend()
    plt.title(u"Nombre de cécidomyie au cours du temps")
    plt.show()
    
    plt.plot(range(xJours[0],xJours[-1]+1),I_simu,'o-',label=u"Simulé")
    plt.plot(xJours,I_obs,'o-',label=u"Observé")
    plt.legend()
    plt.title(u"Nombre d'inflos au desus des pièges au cours du temps")
    plt.show()
    
    plt.plot(np.log(distvalues))
    plt.legend()
    plt.title("Fonction objectif")
    plt.show()
    print min(distvalues)
    
    print(res)
    
    print (L_simu)
    
    return res 
```
