## Discussion sur la cinétique des larves et des inflos


### BLOC1


#### Larves piégées

Bache : Il y a un pic de larves piégées autour du 20 septembre.

Enh_haut : Il y a aussi un pic de larves piégées autour du 20 septembre mais ça augmente un peu avant. 

Enh_ras : A partir de mi-aout, le nombre de larves piégées augmente, la cinétique est en "dents de scie" jusqu'au pic du 20 septembre avec une diminution importante fin août.  


#### Inflos vivantes

Bache : Le nombre d'inflorescences augmentent à partir de fin août et reste élevé jusqu'à fin septembre, puis commence à diminuer début octobre.

Enh_haut : Cinétique similaire sauf commence à diminuer plus tôt (fin septembre), mais beaucoup plus d'inflos en général.

Enh_ras : Cinétique très différente, le nombre d'inflorescences augmentent dès le début pour atteindre un pic à la mi-août, puis diminue rapidement, puis lentement jusqu'à la fin. 


#### Inflos mortes 

Bache : A partir de début août, le nombre d'inflos mortes augmente régulièrement, puis de façon exponentielle à partir de fin septembre.

Enh_haut : Cinétique similaire, avec des dents de scie plus importantes.

Enh_ras : Le nombre d'inflos mortes augmente brutalement à partir de fin août, puis reste élevé tout en faisant des dents de scie.


#### INTERPRETATION

Bache : Dans cette parcelle, le pic de floraison est de fin août à fin septembre. Le nombre de larves capturées avant est très faible à cause de la couverture du sol qui empêche les larves en diapause de l'année précédente d'émerger. Puis, il est très fort lors du pic de floraison et chute à cause de la diminution du nombre d'inflos vivantes (fin de floraison) et de l'augmentation du nombre d'inflos mortes (par les cécidomyies). 

Enh_haut : On peut faire les mêmes commentaires que précédemment, sauf le nombre de larves piégées qui augmente avant le pic dûe aux larves diapausantes émergeant.

Enh_ras : Le pic de floraison est décalé par rapport aux parcelles précédentes, il est plus tôt, vers la mi-août, soit un mois plus tôt. Et donc le nombre d'inflos mortes augmente plus tôt aussi. On remarque que dès le début le nombre de larves piégées est plus important que dans les deux cas précédents, ce qui confirme l'idée que les larves diapausantes ont plus de facilité à émerger avec un sol nu.


### BLOC 2


#### Larves piégées

Bache : Le nombre de larves piégées augmente très rapidement pour atteindre un pic au 23 août et diminue brutalement. 

Enh_haut : Cinétique similaire mais augmente plus lentement au début. 

Enh_ras : Le nombre de larves piégées reste nul jusqu'au pic du 6 septembre et diminue presque aussi rapidement. 


#### Inflos vivantes

Bache : Le nombre d'inflorescences augmente dès le début. La période intense de floraison dure jusqu'à début septembre.

Enh_haut : Cinétique similaire mais moins d'inflorescences. 

Enh_ras : Cinétique très différentes, pas d'inflos jusqu'à mi-août, puis augmente sensiblement avant le pic du 6 septembre, période de floraison très courte. 


#### Inflos mortes

Bache : Le nombre d'inflos mortes augmente graduellement jusqu'au 20 août, puis diminue. 

Enh_haut : Le nombre d'inflos mortes augmente à partir de mi-août, puis reste en dents de scie. 

Enh_ras : Les inflos meurent seulement à partir du 6 septembre et de façon exponentielle. Elles sont beaucoup plus nombreuses que dans les parcelles précédentes.


#### INTERPRETATION

Bache : La période de floraison est au début, donc logiquement le nombre d'inflos est important, puis diminue après, à l'inverse du nombre d'inflos mortes et les larves piégées sont nombreuses au début et moindre à la fin. 

Enh_haut : Mêmes commentaires

Enh_ras : Pas de larves piégées au début, ni d'inflos vivantes, donc s'il y a des larves diapausantes qui émergent, elles ne peuvent pas trouver d'habitat donc meurent. Les larves augmentent avec le pic de floraison. Le nombre d'inflos qui meurent est très important due à l'attaque de cécidomyies juste avant sûrement. 


### POURQUOI PLUS DE LARVES CAPTUREES A LA FIN ALORS QU'IL Y A TOUJOURS DES INFLOS?

On peut l'expliquer par l'augmentation des températures à la fin, qui augmente le temps de diapause d'après le papier de Paul. Les larves émergent "plus" avec des températures fraiches


### TEMPERATURES RELEVEES

D'après le raport de Lucie, il y a trois périodes distinctes :
- De début juillet à début août, les températures sont constantes avec une faible amplitude journalière. 
- De début août à début octobre, les températures sont constantes avec une plus grande amplitude journalière. 
- De début octobre à fin novembre : augmentation nette des températures.
