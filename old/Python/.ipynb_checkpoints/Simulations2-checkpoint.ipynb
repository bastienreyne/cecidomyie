# Simulations 

Dans ce fichier, nous avons les simulations avec R qui vaut : 
$$R =  \begin{cases} 1  & \text{ si } N_{t-1} < k \times I_{t-1} \\  \frac{k \times I_{t-1}}{N_{t-1}} & \text{ sinon. }  \end{cases} $$

On ne prend pas en compte la durée de développement des larves.

On fixe $\lambda = 150$, $\mu_{MS}p_l = 0.5$ et k=1.
