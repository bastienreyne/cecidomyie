## Fonctions

```python
# Pour la lecture des fichiers 

def bloc (Nom, groupe, Sheet, un, deux, trois, quatre, cinq) :
    f = read_excel('../Fichiers de donnees/Donnees re organisees/Fichier donnees floraison/'+Nom+'.xls')
    liste = f.groupby(groupe).sum()
    nouvelles = liste[un]
    vivantes = liste[deux]
    mortes = liste[trois]
    vivantes_theo = liste[quatre]
    mortes_theo = liste[cinq]  
    return nouvelles, vivantes, mortes, vivantes_theo, mortes_theo


def plot_ (Jours, nouvelles, vivantes, mortes, vivantes_theo, mortes_theo) :
    plt.plot(Jours,vivantes,label="Vivantes")
    plt.plot(Jours,nouvelles,label="Nouvelles")
    plt.plot(Jours,mortes,label="Mortes")
    plt.legend()
    plt.xticks(rotation=60)
    plt.title(u"Données relevées")
    plt.show()
    plt.plot(Jours,vivantes_theo,label="Vivantes")
    plt.plot(Jours,nouvelles,label="Nouvelles")
    plt.plot(Jours,mortes_theo,label="Mortes")
    plt.legend()
    plt.xticks(rotation=60)
    plt.title(u"Données théoriques")
    plt.show()
    plt.plot(Jours,vivantes_theo,label="Vivantes theoriques")
    plt.plot(Jours,vivantes,label=u"Vivantes observées")
    plt.plot(Jours,vivantes_theo-vivantes,label="Difference")
    plt.xticks(rotation=60)
    plt.legend()
    plt.title(u"Comparaison")
    plt.show()
    
    
def fichier_piege (Nom) :
    f = read_excel('../Fichiers de donnees/Donnees re organisees/Fichier piege/'+Nom+'.xls')
    df = f.groupby('date').sum()
    nb_larves = df["larves"]
    nb_inflo = df["inflos_vivantes"]
    nb_inflo_morte = df["inflos_mortes"]
    return nb_larves, nb_inflo, nb_inflo_morte


def fichier_donnees_floraison (Nom) :
    f = read_excel('../Fichiers de donnees/Donnees re organisees/Fichier donnees floraison/'+Nom+'.xls')
    df = f.groupby('date').sum()
    nouvelles = df["nouvelles"]
    vivantes = df["vivantes"]
    mortes = df["mortes"]
    return nouvelles, vivantes, mortes
```

```python
# Pour l'optimisation

def integrate (a, inflos_donnees_floraison) :
    inflos_interpo = np.zeros(len(inflos_donnees_floraison))
    for i in range (len(inflos_donnees_floraison)) :
        inflos_interpo[i] = a*inflos_donnees_floraison[i]
    return inflos_interpo

def distance (Simu,Obs) :
    return np.sqrt(sum((Simu-Obs)**2))

def objectif (params, inflos_donnees_floraison, inflos_donnees_piege) :
    val = distance(integrate(params, inflos_donnees_floraison), inflos_donnees_piege)
    return val

def visual_optimization_bis (date, inflos_donnees_floraison, inflos_donnees_piege, p0) :  
    min_ = 99999999999999999999999
    N = len(p0)
    for cpt in range (N):
        i  = p0[cpt]
        b = least_squares(objectif,i,args=(inflos_donnees_floraison, inflos_donnees_piege)).cost
        if (b<min_):
            min_ = b
            min_init = i
    res = least_squares(objectif,min_init,args=(inflos_donnees_floraison, inflos_donnees_piege))
    x_ = res.x
    cost_ = res.cost
    y = integrate(x_, inflos_donnees_floraison)
    plt.plot(date, y, label="Inflos_floraison * a")
    plt.plot(date, inflos_donnees_piege, label="Inflos_piege")
    plt.legend()
    plt.xticks(rotation=60)
    plt.show()
    return x_
```

```python
# Pour être sur la même période entre les données du fichier piège et celles du fichier donnees.floraison

def debut_fin (donnees_foraison, donnees_piege):
    date_1 = donnees_foraison.index
    date_2 = donnees_piege.index
    debut = date_1[0]
    if (debut.replace(year=2017)<=date_2[0]) :
        debut = date_2[0]
    fin = date_1[np.shape(date_1)[0]-1]
    if (fin.replace(year=2017)>=date_2[np.shape(date_2)[0]-1]) :
        fin = date_2[np.shape(date_2)[0]-1]
    for i in range (np.shape(date_1)[0]) :
        if (debut.replace(year=2017)==date_1[i].replace(year=2017)) :
            ii = i
        if (fin.replace(year=2017)==date_1[i].replace(year=2017)) :
            jj = i
    for i in range (np.shape(date_2)[0]) :
        if (debut.replace(year=2017)==date_2[i].replace(year=2017)) :
            kk = i
        if (fin.replace(year=2017)==date_2[i].replace(year=2017)) :
            ll = i
    date = date_2[kk:ll]
    return (date, ii,jj,kk,ll)
```

# Avec les tranches d'âge

Voir https://interstices.info/jcms/i_56750/modeliser-la-dynamique-des-populations-animales-la-predation

Comme précédemment, on garde le nombre de nouvelles inflorescences à chaque date d'observation. 

On note N_CD, N_E, N_F et N_G, respectivement le nombre d'inflorescences aux stades C/D, E, F et G. Ce sont des vecteurs contenant l'information pour chaque jour du début à la fin des observations, ils sont de taille la différence entre le dernier et le premier jour des observations. 

Pour N_CD, on reprend le nombre de nouvelles inflorescences fourni par les données pour chaque date d'observation. On met à 0, le nombre de nouvelles inflorescences pour les jours où il n'y a pas d'observation. 

On utilise les durées de stade de la thèse de Paul qui paraissent cohérentes face aux graphes précédents.

Durées des stades :  [C : 1 jour, D : 7 jours, E : 7 jours, F : 28 jours, G : 7 jours]

On utilise les équations suivantes : 

$N_E[t] = N_E[t-1] + \frac{N_{CD}[t-1]}{\lambda _{CD}} - \frac{N_{E}[t-1]}{\lambda _{E}} - m_EN_E[t-1]$

$N_F[t] = N_F[t-1] + \frac{N_{E}[t-1]}{\lambda _{E}} - \frac{N_{F}[t-1]}{\lambda _{E}} - m_FN_F[t-1]$

$N_G[t] = N_G[t-1] + \frac{N_{F}[t-1]}{\lambda _{F}} - \frac{N_{G}[t-1]}{\lambda _{G}} - m_GN_G[t-1]$

où $\lambda _i$ est la durée du stade $i$ et $m_i$ est son taux de mortalité.

## Dans un premier temps, sans la mortalité

```python
n = len(Bloc1_nouvelles)
nn = (Bloc1_Jours[-1]-Bloc1_Jours[0]).days
a = to_datetime(np.linspace(0,nn,nn+1),origin=Timestamp(Bloc1_Jours[0]),unit='D')
z = np.array(Bloc1_nouvelles)
N_CD = np.zeros(nn)
ii = np.zeros(n)
for i in range (n):
    j = 0
    while (Bloc1_Jours[i]!=a[j]):
        j=j+1
    ii[i] = j
for i in range (n-1):
    N_CD[int(ii[i])] = z[i]
N_E = np.zeros(nn)
N_F = np.zeros(nn)
N_G = np.zeros(nn)
duree = np.array([9,7,28,7])
```

```python
N_E[1] = N_CD[0]/duree[0]
for k in range(2,nn):
    N_E[k] = N_CD[k-1]/duree[0] + N_E[k-1]*(1-1/duree[1]) 

N_F[1] = N_E[0]/duree[1]
for k in range(2,nn):
    N_F[k] = N_E[k-1]/duree[1] + N_F[k-1]*(1-1/duree[2])

N_G[1] = N_F[0]/duree[2]
for k in range(2,nn):
    N_G[k] = N_F[k-1]/duree[2] + N_G[k-1]*(1-1/duree[3])
    
    
N_CD_obs = z
N_E_obs = np.zeros(n)
N_F_obs = np.zeros(n)
N_G_obs = np.zeros(n)

for i in range (n-1):
    N_E_obs[i] = N_E[int(ii[i])]
    N_F_obs[i] = N_F[int(ii[i])]
    N_G_obs[i] = N_G[int(ii[i])]
    
N_E_obs[n-1] = N_E[nn-1]
N_F_obs[n-1] = N_F[nn-1]
N_G_obs[n-1] = N_G[nn-1]
```

```python
N_Stades_obs = N_CD_obs+N_E_obs+N_F_obs+N_G_obs
plt.plot(Bloc1_Jours, N_Stades_obs)
plt.plot(Bloc1_Jours,Bloc1_vivantes)
plt.xticks(rotation=60)
plt.show()
```

D'abord, comme nous n'avons pas introduit de mortalité, le nombre d'inflorescences vivantes ne cesse de croître avec la méthode des tranches d'âge. Nous voyons sur le graphique que le nombre maximal d'inflorescences vivantes observé (environ 200) est obtenu au même moment (1er septembre) qu'avec les tranches d'âge.
