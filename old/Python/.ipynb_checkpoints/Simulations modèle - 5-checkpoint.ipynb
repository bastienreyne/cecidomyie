```python
# Cette fonction renvoie chaque variable du système au temps t suivant. 

def step (t, NbLarvesTroisiemeStade_enh_haut, NbLarvesTroisiemeStade_enh_ras, NbLarvesTroisiemeStade_bache, Nt_enh_haut, Lt_enh_haut, Lt_p_enh_haut, R_enh_haut, Nt_enh_ras, Lt_enh_ras, Lt_p_enh_ras, R_enh_ras, Nt_bache, Lt_bache, Lt_p_bache, R_bache, a_1, a_2, a_3, a_4, RR=0) :
    
    # Enherbement haut
    
    if (t <= d_l) : 
        R = 1
    else :
        if (Nt_enh_haut[t-1] < It_enh_haut[t-1]):
            R = 1
        else :
            R = k*It_enh_haut[t-1]/Nt_enh_haut[t-1]
    
    if (t <= d_l) :
        Lt_plus_1 = 0
    else : 
        Lt_plus_1 = Nt_enh_haut[-1-d_l] * R * E * mu / 2 # On divise par 2 à cause du sex ratio
    
    Lt_p_plus_1 = Lt_plus_1 * beta
    
    Nt_plus_1 = 150
    if (t > d_p) :
        Nt_plus_1 +=  Lt_enh_haut[-1-d_p] * 0.5 * p_pup + Lt_enh_ras[-1-d_p] * 0.5 * p_pup * a_1
    
    if (t+d_l<=TotJours) :
        NbLarvesTroisiemeStade_enh_haut[t+d_l] = Nt_plus_1 * R * E * mu / 2 
    
    Nt_enh_haut.append(Nt_plus_1)
    Lt_enh_haut.append(Lt_plus_1)
    Lt_p_enh_haut.append(Lt_p_plus_1)
    R_enh_haut.append(R)
    
    # ------------------------------------------------------------------ #
    
    # Enherbement ras
    
    if (t <= d_l) : 
        R = 1
    else :
        if (Nt_enh_ras[t-1] < It_enh_ras[t-1]):
            R = 1
        else :
            R = k*It_enh_ras[t-1]/Nt_enh_ras[t-1]
            
    if (t <= d_l) :
        Lt_plus_1 = 0
    else : 
        Lt_plus_1 = Nt_enh_ras[-1-d_l] * R * E * mu / 2 # On divise par 2 à cause du sex ratio
    
    
    Lt_p_plus_1 = Lt_plus_1 * beta
    
    Nt_plus_1 = 150
    if (t > d_p) :
        Nt_plus_1 +=  Lt_enh_ras[-1-d_p] * 0.5 * p_pup + Lt_enh_haut[-1-d_p] * 0.5 * p_pup * a_2
    
    Nt_enh_ras.append(Nt_plus_1)
    Lt_enh_ras.append(Lt_plus_1)
    Lt_p_enh_ras.append(Lt_p_plus_1)
    R_enh_ras.append(R)
    
    # ------------------------------------------------------------------ #
    
    # Bache 
    
    if (t <= d_l) : 
        R = 1
    else :
        if (Nt_bache[t-1] < It_bache[t-1]):
            R = 1
        else :
            R = k*It_bache[t-1]/Nt_bache[t-1]
        
    if (t <= d_l) :
        Lt_plus_1 = 0
    else : 
        Lt_plus_1 = Nt_bache[-1-d_l] * R * E * mu / 2 # On divise par 2 à cause du sex ratio
    
    Lt_p_plus_1 = Lt_plus_1 * beta
    
    Nt_plus_1 = 150
    if (t > d_p) :
        Nt_plus_1 +=  Lt_enh_ras[-1-d_p] * 0.5 * p_pup * a_3 + Lt_enh_haut[-1-d_p] * 0.5 * p_pup * a_4
    
    Nt_bache.append(Nt_plus_1)
    Lt_bache.append(Lt_plus_1)
    Lt_p_bache.append(Lt_p_plus_1)
    R_bache.append(R)
    
    return NbLarvesTroisiemeStade_enh_haut, NbLarvesTroisiemeStade_enh_ras, NbLarvesTroisiemeStade_bache, Nt_enh_haut, Lt_enh_haut, Lt_p_enh_haut, R_enh_haut, Nt_enh_ras, Lt_enh_ras, Lt_p_enh_ras, R_enh_ras, Nt_bache, Lt_bache, Lt_p_bache, R_bache
```

```python
# Cette fonction calcule toutes les variables du système au fil du temps et renvoie Lt_p, le nombre de larves piégées à chaque
# pas de temps. C'est sur ces données qu'on fait la minimisation.

def integrate (a_1, a_2, a_3, a_4, RR=0) :
    
    # Initialisation 
    NbLarvesTroisiemeStade_enh_haut = np.zeros(TotJours+1)
    NbLarvesTroisiemeStade_enh_ras = np.zeros(TotJours+1)
    NbLarvesTroisiemeStade_bache = np.zeros(TotJours+1)
    
    Nt_enh_haut = [150]
    Lt_enh_haut = [0]
    Lt_p_enh_haut = [0]
    R_enh_haut = []
    
    Nt_enh_ras = [150]
    Lt_enh_ras = [0]
    Lt_p_enh_ras = [0]
    R_enh_ras = []
    
    Nt_bache = [150]
    Lt_bache = [0]
    Lt_p_bache = [0]
    R_bache = []
    
    for i in range (1,TotJours+1) :
        NbLarvesTroisiemeStade_enh_haut, NbLarvesTroisiemeStade_enh_ras, NbLarvesTroisiemeStade_bache, Nt_enh_haut, Lt_enh_haut, Lt_p_enh_haut, R_enh_haut, Nt_enh_ras, Lt_enh_ras, Lt_p_enh_ras, R_enh_ras, Nt_bache, Lt_bache, Lt_p_bache, R_bache = step (i, NbLarvesTroisiemeStade_enh_haut, NbLarvesTroisiemeStade_enh_ras, NbLarvesTroisiemeStade_bache, Nt_enh_haut, Lt_enh_haut, Lt_p_enh_haut, R_enh_haut, Nt_enh_ras, Lt_enh_ras, Lt_p_enh_ras, R_enh_ras, Nt_bache, Lt_bache, Lt_p_bache, R_bache, a_1, a_2, a_3, a_4, RR)

    return np.array(Lt_enh_haut), np.array(Lt_p_enh_haut), np.array(Nt_enh_haut), np.array(R_enh_haut), np.array(Lt_enh_ras), np.array(Lt_p_enh_ras), np.array(Nt_enh_ras), np.array(R_enh_ras), np.array(Lt_bache), np.array(Lt_p_bache), np.array(Nt_bache), np.array(R_bache) 
```

```python
distvalues = []

# Cette fonction minimise sur le nombre de larves piégées
def objectif1_1 (params) :
    global distvalues
    val = distance1(integrate(*params)[1], nb_larves_enh_haut) + distance1(integrate(*params)[5], nb_larves_enh_ras) + distance1(integrate(*params)[9], nb_larves_bache) + abs(sum(integrate(*params)[1])-sum(nb_larves_enh_haut)) + abs(sum(integrate(*params)[5])-sum(nb_larves_enh_ras)) + abs(sum(integrate(*params)[9])-sum(nb_larves_bache))
    distvalues.append(val)
    return val

def objectif1_2 (params) :
    global distvalues
    val = distance2(integrate(*params)[1], nb_larves) + abs(sum(integrate(*params)[1])-sum(nb_larves))
    distvalues.append(val)
    return val

def objectif1_Inf (params) :
    global distvalues
    val = distanceInf1(integrate(*params)[1], nb_larves) #+ abs(sum(integrate(*params)[1])-sum(nb_larves))
    distvalues.append(val)
    return val
```

```python
# Fonctions distance pour les fonctions objectifs 

# Distance 1
def distance1 (Simu,Obs) :
    return sum(abs(Simu-Obs))

# Distance euclidienne
def distance2 (Simu,Obs) :
    return np.sqrt(sum((Simu-Obs)**2))

# Distance inf 
def distanceInf1 (Simu,Obs) :
    return max(abs(Simu-Obs))

def distanceInf2 (Simu1,Obs1,Simu2,Obs2) :
    return max(max(abs(Simu1-Obs1)),max(abs(Simu2-Obs2)))
```

```python
# Fonctions d'optimisation

def optimize1 (p0) :
    res = least_squares(objectif1_1, p0, bounds=bounds, method='trf')
    return res.x

def optimize2 (p0) :
    res = least_squares(objectif1_2, p0, bounds=bounds, method='trf')
    return res.x

def optimizeInf (p0) :
    res = least_squares(objectif1_Inf, p0, bounds=bounds, method='trf')
    return res.x
```
