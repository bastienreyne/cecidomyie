# Simulations 

Dans ce fichier, nous avons les simulations avec R qui vaut : 
$$R =  \begin{cases} 1  & \text{ si } N_{t-1} < k \times I_{t-1} \\  \frac{k \times I_{t-1}}{N_{t-1}} & \text{ sinon. }  \end{cases} $$

On ne prend pas en compte la durée de développement des larves.

Il y a estimation de $\lambda$, $\mu_{MS}$ et $p_l$.

## Importation des librairies

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from zFonctionsGenerales import *
from zFonctionsSimulation1 import *
```

## Paramètres pour optimisation

```python
# Bornes pour les paramètres à estimer
bounds = ([0, 0, 0], [np.inf, 1, 1])
```

```python
lambda_0 = np.arange(100,1000,400)
mu_MS0 = np.arange(0.1,1,0.4)
p_l0 = np.arange(0.1,1,0.4)
p0 = (lambda_0, mu_MS0, p_l0)
```
