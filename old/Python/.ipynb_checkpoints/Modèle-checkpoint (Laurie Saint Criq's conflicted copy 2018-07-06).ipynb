### Loi de développement du système

**Soit $N_t$ la population estimée d'adultes dans le verger à la date t.** Pour simuler la dynamique de cette population, on utilise l'équation suivante:

$$N_{t+1} = \lambda + L_{t+1-d_p} \times \mu_{MS} \times p_{pup} \times p_{l} $$

avec $\mu_{MS} = \mu^1_{MS} \times \mu^2_{MS}$.

Cette équation signifie que la population d'adultes dans le verger à la date $t+1$ et à la date $t-1$ est le nombre d'adultes entrant dans le verger ($\lambda$), plus le nombre de larves qui émergent du sol entrées en pupaison à la date $t+1-d_p$.

* $\lambda_t$ est le nombre d'adultes entrant dans le verger à la date t. Pour l'instant c'est un nombre constant au cours du temps appelé $\lambda$.

* $\mu^1_{MS}$ est la probabilité de survie de la larve de l’éjection au sol jusqu’à ce qu’elle rentre en terre selon la modalité du sol.

* $\mu^2_{MS}$ est la probabilité pour l’adulte de sortir du sol selon la modalité du sol.

* $p_{l}$ est la probabilité pour l’adulte de rester dans le verger quand il sort du sol. 

*  $d_p$ est la durée de pupaison. Entre 4 et 6 jours.

**Soit $L_t$ le nombre estimé de larves qui s'éjectent à la date t.**

$$L_{t} = N_{t-d_l} \times R \times E \times \mu$$

* $E$ est le nombre maximal d'oeufs pondus par une femelle.
* $\mu$ est la probabilité de survie des oeufs jusqu'au troisième stade de larve.
* $R$ la disponibilité en ressource (inflorescences).
* $d_l$ est la durée de dévelopement des larves et des oeufs. Entre 7 et 12 jours.


**Soit $L^p_t$ est le nombre estimé de larves piégées à la date t.**

$$L^p_t = L_t \times \beta$$ 

* $\beta$ est l'efficacité du piège.
   - Les pièges utilisés en 2017 sont différents de ceux utilisés dans l'expérience de Paul. Il faudrait réestimer l'efficacité de ce nouveau type de pièges.
   - La valeur du beta dans le modèle de Paul est > 1. Ce qui parait un peu abérant (plus de larves piégées que de larves qui tombent).

**Soit $I^d_t$ la population d'inflos à l'instant $t$ apparues le jour $d$.**

$$I^d_{t} = \begin{cases}
I^+(d) & \text{si } t = d \\
I^d_{t-1} (1 - \gamma \times N_{t-1}) & \text{si } d < t < d+T\\
0 & \text{sinon }
\end{cases}$$

 * $T$ la durée de vie d'une inflorescence.
 
Soit $I_t$ le nombre estimé d'inflos dans le verger.

$$I_t = \sum_d I^d_t$$

**Pour le moment, $I_t$ est une entrée du modèle. On le récupère des données de Christian et Alain**


On estime la disponibilité en ressource (inflorescence) par un coefficient $R$ estimé de la manière suivante:

$$R =  \begin{cases} 1  & \text{ si } N_t < k \times I_t \\  \frac{k \times I_t}{N_t} & \text{ sinon. }  \end{cases} $$

 * $k$ est une constante. C'est le nombre de piqûre qui garantit la mort de l'inflo. 

Ca sert à ne pas compter les larves si elles ne peuvent pas se poser sur une inflo (i.e. s'il n'y a pas de ressources dans le verger)


#### Liste des paramètres recherchés

$\lambda_t$ est le nombre d'adultes entrant dans le verger à la date t. Pour l'instant c'est un nombre constant au cours du temps appelé $\lambda$.

$\mu^1_{MS}$ est la probabilité de survie de la larve de l’éjection au sol jusqu’à qu’elle rentre en terre selon la modalité du sol

$\mu^2_{MS}$ est la probabilité pour l’adulte de sortir du sol selon la modalité du sol

$p_{l}$ est la probabilité pour l’adulte de rester dans le verger quand il sort du sol selon la modalité du sol

$k$ est le nombre de piqûres qui garantit la mort de l'inflo

$\gamma$ est le taux de dégradation des inflos 

#### Liste des paramètres connus

$\bar{L_t}$ est le nombre observé de larves piégées à la date t.

$E$ est le nombre maximal d'oeufs pondus par une femelle.

($I^+$ est le nombre d'apparition d'inflos.)

$d_l$ est la durée de dévelopement des larves et des oeufs. Entre 7 et 12 jours.

$d_p$ est la durée de pupaison. Entre 4 et 6 jours.

$\beta$ est l'efficacité du piège. 

$p_{pup}$ est la probabilité d'entrer en pupaison.

$\mu$ est la probabilité de survie des oeufs jusqu'au troisième stade de larve.
