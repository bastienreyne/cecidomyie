```python
def fichier (Nom, groupe, Sheet) :
    
    f = read_excel(Nom+'.xls', Sheet)
    df = f.groupby(groupe).sum()
    
    return df
```

```python
def step (t, Nt_enh_haut, Lt_enh_haut, R_enh_haut, NbAdultesEmergents_enh_haut, NbLarvesTroisiemeStade_enh_haut, Nt_enh_ras, Lt_enh_ras, R_enh_ras, NbAdultesEmergents_enh_ras, NbLarvesTroisiemeStade_enh_ras, Nt_bache, Lt_bache, R_bache, NbAdultesEmergents_bache, NbLarvesTroisiemeStade_bache, lambda_, mu_MS_pl_enh_haut, mu_MS_pl_enh_ras, k, a_enh_haut, a_enh_ras, a_bache, RR=0):
    
    a = a_enh_haut + a_enh_ras + a_bache
    
    # ENHERBEMENT HAUT 
    
    Nt_plus_1_enh_haut = int(round(lambda_*a_enh_haut/a + NbAdultesEmergents_enh_haut[t]*0.5))
    if (Nt_plus_1_enh_haut < It_enh_haut[t-1]):
        R_plus_1_enh_haut = 1
    else : 
        R_plus_1_enh_haut = k*It_enh_haut[t-1]/Nt_plus_1_enh_haut
        
    NbLarves = np.sum(np.random.poisson(E*mu,int(round(Nt_plus_1_enh_haut*R_plus_1_enh_haut/2))))
    
    DureeDvpt = np.random.poisson(d_l,NbLarves)
    
    for i in DureeDvpt :
        if (t+i+1<=TotJours) :
            NbLarvesTroisiemeStade_enh_haut[t+i+1] += 1
    
    NbLarves = int(round(NbLarves*mu_MS_pl_enh_haut))
    
    NbLarvesPupaison = np.sum(np.random.uniform(0,1,NbLarves)<p_pup)
    
    DureePup = np.random.poisson(d_p,NbLarvesPupaison)
    
    for i in DureePup :
        if (t+i+1<=TotJours) :
            NbAdultesEmergents_enh_haut[t+i+1] += 1
    
    Lt_plus_1_enh_haut = NbLarvesTroisiemeStade_enh_haut[t]
    
    Nt_enh_haut.append(Nt_plus_1_enh_haut)
    Lt_enh_haut.append(Lt_plus_1_enh_haut)
    R_enh_haut.append(R_plus_1_enh_haut)
    
    
    # ENHERBEMENT RAS
    
    Nt_plus_1_enh_ras = int(round(lambda_*a_enh_ras/a + NbAdultesEmergents_enh_ras[t]))
    if (Nt_plus_1_enh_ras < It_enh_ras[t-1]):
        R_plus_1_enh_ras = 1
    else : 
        R_plus_1_enh_ras = k*It_enh_ras[t-1]/Nt_plus_1_enh_ras    
    
    NbLarves = np.sum(np.random.poisson(E*mu,int(round(Nt_plus_1_enh_ras*R_plus_1_enh_ras/2))))
    
    DureeDvpt = np.random.poisson(d_l,NbLarves)
    
    for i in DureeDvpt :
        if (t+i+1<=TotJours) :
            NbLarvesTroisiemeStade_enh_ras[t+i+1] += 1
    
    NbLarves = int(round(NbLarves*mu_MS_pl_enh_ras))
    
    NbLarvesPupaison = np.sum(np.random.uniform(0,1,NbLarves)<p_pup)
    
    DureePup = np.random.poisson(d_p,NbLarvesPupaison)
    
    for i in DureePup :
        if (t+i+1<=TotJours) :
            NbAdultesEmergents_enh_ras[t+i+1] += 1
            
    Lt_plus_1_enh_ras = NbLarvesTroisiemeStade_enh_ras[t]
    
    Nt_enh_ras.append(Nt_plus_1_enh_ras)
    Lt_enh_ras.append(Lt_plus_1_enh_ras)
    R_enh_ras.append(R_plus_1_enh_ras)
    
    # BACHAGE 
    
    Nt_plus_1_bache = int(round(lambda_*a_bache/a))
    if (Nt_plus_1_bache < It_bache[t-1]):
        R_plus_1_bache = 1
    else : 
        R_plus_1_bache = k*It_bache[t-1]/Nt_plus_1_bache
        
    NbLarves = np.sum(np.random.poisson(E*mu,int(round(Nt_plus_1_bache*R_plus_1_bache/2))))
    
    DureeDvpt = np.random.poisson(d_l,NbLarves)
    
    for i in DureeDvpt :
        if (t+i+1<=TotJours) :
            NbLarvesTroisiemeStade_bache[t+i+1] += 1
            
    Lt_plus_1_bache = NbLarvesTroisiemeStade_bache[t]
    
    Nt_bache.append(Nt_plus_1_bache)
    Lt_bache.append(Lt_plus_1_bache)
    R_bache.append(R_plus_1_bache)
    
    return Nt_enh_haut, Lt_enh_haut, R_enh_haut, NbAdultesEmergents_enh_haut, NbLarvesTroisiemeStade_enh_haut, Nt_enh_ras, Lt_enh_ras, R_enh_ras, NbAdultesEmergents_enh_ras, NbLarvesTroisiemeStade_enh_ras, Nt_bache, Lt_bache, R_bache, NbAdultesEmergents_bache, NbLarvesTroisiemeStade_bache
```

```python
def integrate (lambda_, mu_MS_pl_enh_haut, mu_MS_pl_enh_ras, k, a_enh_haut, a_enh_ras, a_bache, RR=0) :
    
    a = a_enh_haut + a_enh_ras + a_bache
    
    Nt_enh_haut = [int(round(lambda_*a_enh_haut/a))]
    Lt_enh_haut = [0]
    R_enh_haut = []
    NbAdultesEmergents_enh_haut = np.zeros((TotJours+1))
    NbLarvesTroisiemeStade_enh_haut = np.zeros((TotJours+1))
    
    Nt_enh_ras = [int(round(lambda_*a_enh_ras/a))]
    Lt_enh_ras = [0]
    R_enh_ras = []
    NbAdultesEmergents_enh_ras = np.zeros((TotJours+1))
    NbLarvesTroisiemeStade_enh_ras = np.zeros((TotJours+1))
    
    Nt_bache = [int(round(lambda_*a_bache/a))]
    Lt_bache = [0]
    R_bache = []
    NbAdultesEmergents_bache = np.zeros((TotJours+1))
    NbLarvesTroisiemeStade_bache = np.zeros((TotJours+1))
    
    for i in range (1,TotJours+1) :
        Nt_enh_haut, Lt_enh_haut, R_enh_haut, NbAdultesEmergents_enh_haut, NbLarvesTroisiemeStade_enh_haut, Nt_enh_ras, Lt_enh_ras, R_enh_ras, NbAdultesEmergents_enh_ras, NbLarvesTroisiemeStade_enh_ras, Nt_bache, Lt_bache, R_bache, NbAdultesEmergents_bache, NbLarvesTroisiemeStade_bache = step (i, Nt_enh_haut, Lt_enh_haut, R_enh_haut, NbAdultesEmergents_enh_haut, NbLarvesTroisiemeStade_enh_haut, Nt_enh_ras, Lt_enh_ras, R_enh_ras, NbAdultesEmergents_enh_ras, NbLarvesTroisiemeStade_enh_ras, Nt_bache, Lt_bache, R_bache, NbAdultesEmergents_bache, NbLarvesTroisiemeStade_bache, lambda_, mu_MS_pl_enh_haut, mu_MS_pl_enh_ras, k, a_enh_haut, a_enh_ras, a_bache, RR)
        
    return np.array(Nt_enh_haut), np.array(Lt_enh_haut), np.array(R_enh_haut), np.array(NbAdultesEmergents_enh_haut), np.array(NbLarvesTroisiemeStade_enh_haut), np.array(Nt_enh_ras), np.array(Lt_enh_ras), np.array(R_enh_ras), np.array(NbAdultesEmergents_enh_ras), np.array(NbLarvesTroisiemeStade_enh_ras), np.array(Nt_bache), np.array(Lt_bache), np.array(R_bache), np.array(NbAdultesEmergents_bache), np.array(NbLarvesTroisiemeStade_bache)
```

```python
distvalues = []

def objectif1_1 (params) :
    global distvalues
    val = distance1(integrate(*params)[0], nb_larves) + abs(sum(integrate(*params)[0])-sum(nb_larves))
    distvalues.append(val)
    return val

def objectif1_2 (params) :
    global distvalues
    val = distance2(integrate(*params)[0], nb_larves) + abs(sum(integrate(*params)[0])-sum(nb_larves))
    distvalues.append(val)
    return val

def objectif1_Inf (params) :
    global distvalues
    val = distanceInf1(integrate(*params)[0], nb_larves) #+ abs(sum(integrate(*params)[0])-sum(nb_larves))
    distvalues.append(val)
    return val
```

```python
def distance1 (Simu,Obs) :
    return sum(abs(Simu-Obs))

def distance2 (Simu,Obs) :
    return np.sqrt(sum((Simu-Obs)**2))

def distanceInf1 (Simu,Obs) :
    return max(abs(Simu-Obs))

def distanceInf2 (Simu1,Obs1,Simu2,Obs2) :
    return max(max(abs(Simu1-Obs1)),max(abs(Simu2-Obs2)))
```

```python
def optimize1 (p0) :
    res = least_squares(objectif1_1, p0, bounds=bounds, method='trf')
    return res.x

def optimize2 (p0) :
    res = least_squares(objectif1_2, p0, bounds=bounds, method='trf')
    return res.x

def optimizeInf (p0) :
    res = least_squares(objectif1_Inf, p0, bounds=bounds, method='trf')
    return res.x
```
