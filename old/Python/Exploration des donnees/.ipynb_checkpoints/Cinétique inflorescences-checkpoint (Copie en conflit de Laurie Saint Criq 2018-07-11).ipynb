## Estimer le nombre d'inflorescences vivantes

Maintenant, nous allons essayer de simuler le nombre d'inflorescences vivantes à chaque pas de temps dans le verger en connaissant le nombre de nouvelles inflorescences à chaque instant et le nombre de larves à chaque instant. 

Soit $I^d_t$ la population d'inflos présentes à l'instant $t$ apparues le jour $d$.

$$I^d_{t} = \begin{cases}
I^+(d) & \text{si } t = d \\
I^d_{t-1} (1 - \gamma \times N_{t-1}) & \text{si } d < t < d+T\\
0 & \text{sinon }
\end{cases}$$

 * $T$ la durée de vie d'une inflorescence.
 * $\gamma$ est un paramètre à estimer. 
 
Soit $I_t$ le nombre estimé d'inflos dans le verger.

$$I_t = \sum_d I^d_t$$

```python
# BLOC1
I_plus = nouvelles_1_j
N = piege_larves_bloc1_bache_j
I_real = piege_inflo_bloc1_bache_j
tmp = np.shape(I_plus)[0]
T = 61
gamma = 0.5

It_d = np.diag(I_plus)

for d in range(tmp) :
    for t in range ((d+1),min((d+T),tmp)) :
        It_d[t,d] = It_d[t-1,d]
I_sim = np.sum(It_d,1)
print(I_sim)
plt.plot(date_1_j,I_sim)
plt.plot(date_1_j,I_real)
plt.show()
```

# Avec les tranches d'âge

Voir https://interstices.info/jcms/i_56750/modeliser-la-dynamique-des-populations-animales-la-predation

Comme précédemment, on garde le nombre de nouvelles inflorescences à chaque date d'observation. 

On note N_CD, N_E, N_F et N_G, respectivement le nombre d'inflorescences aux stades C/D, E, F et G. Ce sont des vecteurs contenant l'information pour chaque jour du début à la fin des observations, ils sont de taille la différence entre le dernier et le premier jour des observations. 

Pour N_CD, on reprend le nombre de nouvelles inflorescences fourni par les données pour chaque date d'observation. On met à 0, le nombre de nouvelles inflorescences pour les jours où il n'y a pas d'observation. 

On utilise les durées de stade de la thèse de Paul qui paraissent cohérentes face aux graphes précédents.

Durées des stades :  [C : 1 jour, D : 7 jours, E : 7 jours, F : 28 jours, G : 7 jours]

On utilise les équations suivantes : 

$N_E[t] = N_E[t-1] + \frac{N_{CD}[t-1]}{\lambda _{CD}} - \frac{N_{E}[t-1]}{\lambda _{E}} - m_EN_E[t-1]$

$N_F[t] = N_F[t-1] + \frac{N_{E}[t-1]}{\lambda _{E}} - \frac{N_{F}[t-1]}{\lambda _{E}} - m_FN_F[t-1]$

$N_G[t] = N_G[t-1] + \frac{N_{F}[t-1]}{\lambda _{F}} - \frac{N_{G}[t-1]}{\lambda _{G}} - m_GN_G[t-1]$

où $\lambda _i$ est la durée du stade $i$ et $m_i$ est son taux de mortalité.

## Dans un premier temps, sans la mortalité

```python
n = len(Bloc1_nouvelles)
nn = (Bloc1_Jours[-1]-Bloc1_Jours[0]).days
a = to_datetime(np.linspace(0,nn,nn+1),origin=Timestamp(Bloc1_Jours[0]),unit='D')
z = np.array(Bloc1_nouvelles)
N_CD = np.zeros(nn)
ii = np.zeros(n)
for i in range (n):
    j = 0
    while (Bloc1_Jours[i]!=a[j]):
        j=j+1
    ii[i] = j
for i in range (n-1):
    N_CD[int(ii[i])] = z[i]
N_E = np.zeros(nn)
N_F = np.zeros(nn)
N_G = np.zeros(nn)
duree = np.array([9,7,28,7])
```

```python
N_E[1] = N_CD[0]/duree[0]
for k in range(2,nn):
    N_E[k] = N_CD[k-1]/duree[0] + N_E[k-1]*(1-1/duree[1]) 

N_F[1] = N_E[0]/duree[1]
for k in range(2,nn):
    N_F[k] = N_E[k-1]/duree[1] + N_F[k-1]*(1-1/duree[2])

N_G[1] = N_F[0]/duree[2]
for k in range(2,nn):
    N_G[k] = N_F[k-1]/duree[2] + N_G[k-1]*(1-1/duree[3])
    
    
N_CD_obs = z
N_E_obs = np.zeros(n)
N_F_obs = np.zeros(n)
N_G_obs = np.zeros(n)

for i in range (n-1):
    N_E_obs[i] = N_E[int(ii[i])]
    N_F_obs[i] = N_F[int(ii[i])]
    N_G_obs[i] = N_G[int(ii[i])]
    
N_E_obs[n-1] = N_E[nn-1]
N_F_obs[n-1] = N_F[nn-1]
N_G_obs[n-1] = N_G[nn-1]
```

```python
N_Stades_obs = N_CD_obs+N_E_obs+N_F_obs+N_G_obs
plt.plot(Bloc1_Jours, N_Stades_obs)
plt.plot(Bloc1_Jours,Bloc1_vivantes)
plt.xticks(rotation=60)
plt.show()
```

```python
N_Stades = N_CD+N_E+N_F+N_G
plt.plot(N_Stades)
plt.title("Le nombre d'inflorescences vivantes avec tranches d'age")
plt.show()
plt.plot(Bloc1_vivantes)
plt.title("Le nombre d'inflorescences vivantes observe")
plt.xticks(rotation=60)
plt.show()
```

D'abord, comme nous n'avons pas introduit de mortalité, le nombre d'inflorescences vivantes ne cesse de croître avec la méthode des tranches d'âge. Nous voyons sur le graphique que le nombre maximal d'inflorescences vivantes observé (environ 200) est obtenu au même moment (1er septembre) qu'avec les tranches d'âge.
